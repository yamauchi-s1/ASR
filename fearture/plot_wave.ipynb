```python
import wave
import matplotlib.pyplot as plt
import numpy as np
```

```python
wave_file = '../data/wav/BASIC5000_0001.wav'
out_plot = "./plot.png"

with wave.open(wave_file, 'rb') as wav:
    sample_freauency = wav.getframerate() #サンプリング周波数を取得
    sample_size = wav.getsampwidth() #サンプルサイズを取得
    num_channles = wav.getnchannels() #チャンネルを取得
    num_samples = wav.getnframes() #wavデータのサンプル数を取得
    wavform = wav.readframes(num_samples) #wavデータを読み込み
    
print(f"Sampling Frequency  : {sample_freauency}")
print(f"Sample Size         : {sample_size}")
print(f"Number of Channnels : {num_channles}")
print(f"Number of Samples   : {num_samples}")
    
```

```python
wav_data = np.frombuffer(wavform, dtype=np.int16)
time_axis = np.arange(num_samples) / sample_freauency

# プロットの描画領域を作成
plt.figure(figsize=(10,4))

# プロット
plt.plot(time_axis, wav_data)

# 横軸と縦軸のラベルを定義
plt.xlabel("Time [sec]")
plt.ylabel("Value")

# 横軸の表示領域を0から波形終了時刻に制限
plt.xlim([0, num_samples / sample_freauency])
```
